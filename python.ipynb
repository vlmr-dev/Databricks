Escreva um programa que permute o valor de duas variáveis inteiras.

```python
# Leitura das entradas
# Valor a = 5
# Valor b = 8

a = int(input())
b = int(input())

# Permutação dos valores
a, b = b, a

# Exibição dos resultados
print(f"Valor de a após permutação: {a}")
print(f"Valor de b após permutação: {b}")
```

2. Escreva um programa que, a partir de um valor de custo e de um valor de venda, mostre o valor do lucro obtido com a venda do produto.

```python
# Leitura das entradas
# custo = 50
# venda = 80 

custo = float(input())
venda = float(input())

lucro = venda - custo

print(f"O lucro obtido foi R$ {lucro:.2f}.")
```

3. Escreva um programa que receba o nome de uma pessoa e sua idade em anos. O programa deve calcular e exibir a idade da pessoa em meses e dias. Obs.: Considere que um ano tem 12 meses e 365 dias.

```python
# Leitura das entradas
# nome = Maria
# idade_anos = 25

nome = input()
idade_anos = int(input())

idade_meses = idade_anos * 12
idade_dias = idade_anos * 365

print(f"{nome} tem {idade_anos} anos ou {idade_meses} meses ou ainda {idade_dias} dias!")
```

4. Escreva um programa que solicite a altura e o raio de um cilindro. Calcule o volume total do cilindro (use π=3.14) e exiba esse valor.

```python
# Leitura das entradas
# altura = 8
# raio = 2

altura = float(input())
raio = float(input())
pi = 3.14

volume = pi * (raio ** 2) * altura

print(f"O volume do cilindro que tem {altura:.1f} de altura e {raio:.1f} de raio é igual a {volume:.2f}")
```

5. Escreva um programa que receba uma temperatura em Celsius (C) e exiba a temperatura lida usando as escalas Kelvin (K) e Fahrenheit (F). 
Considere as seguintes fórmulas:
K = C + 273
F = 1,8*C + 32

```python
# Leitura das entradas
# temperatura_celsius = 32

temperatura_celsius = float(input())

temperatura_kelvin = temperatura_celsius + 273
temperatura_fahrenheit = 1.8 * temperatura_celsius + 32

print(f"Temperatura em Kelvin: {temperatura_kelvin:.2f}K")

print(f"Temperatura em Fahrenheit: {temperatura_fahrenheit:.2f}°F")
```

6. Escreva um programa que leia as 2 notas de um aluno em uma disciplina. 
Depois exiba quantos pontos o aluno ficou distante da nota 10 para cada avaliação. 
Sua média e quantos pontos a média do aluno ficou distante da nota 10.

```python
# Leitura das entradas
# nota1 = 8
# nota2 = 5

nota1 = float(input())
nota2 = float(input())

dist1 = 10 - nota1
dist2 = 10 - nota2
media = (nota1 + nota2) / 2
dist_media = 10 - media

print(f"Primeira nota: {nota1:.1f}")
print(f"Distância para nota 10: {dist1:.1f}")

print(f"Segunda nota: {nota2:.1f}")
print(f"Distância para nota 10: {dist2:.1f}")

print(f"Média: {media:.1f}")
print(f"Distância para média 10: {dist_media:.1f}")
```

7. Escreva um programa que leia um valor de custo e o percentual de lucro desejado e na sequencia mostre o valor final do produto.

```python
# Leitura das entradas
# custo = 50
# percentual = 3

custo = float(input())
percentual = float(input())

valor_final = custo + (custo * percentual / 100)

print(f"Valor final do produto: R$ {valor_final:.2f}.")
```

8. Escreva um programa que leia o salário de uma pessoa.
Quantas horas ela trabalha por dia e quantos dias ela trabalhou no mês.
Calcule e exiba quanto essa pessoa recebe por hora.

```python
# Leitura das entradas
# salario = 1450
# horas_por_dia = 6
# dias_trabalhados = 20

salario = float(input())
horas_por_dia = int(input())
dias_trabalhados = int(input())

valor_hora = salario / (horas_por_dia * dias_trabalhados)

print(f"A hora-trabalhada tem valor de R$ {valor_hora:.2f}.")
```

9. Escreva um programa que leia:
A quantidade total de ingressos vendidos de um evento.
O percentual de ingressos vendidos como meia-entrada (o restante será inteira).
O valor do ingresso inteiro.
E que calcule e exiba:
A quantidade de ingressos meia-entrada e inteiros vendidos.
O valor faturado com cada tipo de ingresso.
O valor total arrecadado com as vendas.

```python
# Leitura das entradas
# quantidade_total = 5000
# percentual_meia = 20
# valor_inteira = 400

quantidade_total = int(input())
percentual_meia = int(input())
valor_inteira = float(input())

total_meia = quantidade_total * percentual_meia // 100
total_inteira = quantidade_total - total_meia

faturamento_meia = total_meia * (valor_inteira / 2)
faturamento_inteira = total_inteira * valor_inteira
faturamento_total = faturamento_meia + faturamento_inteira

print(f"Quantidade de ingressos meia-entrada vendidos: {total_meia}")
print(f"Quantidade de ingressos inteiros vendidos: {total_inteira}")
print(f"Faturamento com meia-entrada: R${faturamento_meia:.2f}")
print(f"Faturamento com inteira: R${faturamento_inteira:.2f}")
print(f"Faturamento total: R${faturamento_total:.2f}")
```
