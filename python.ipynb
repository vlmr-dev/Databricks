Escreva um programa que permute o valor de duas variáveis inteiras.

```python
# Leitura das entradas
# Valor a = 5
# Valor b = 8

a = int(input())
b = int(input())

# Permutação dos valores
a, b = b, a

# Exibição dos resultados
print(f"Valor de a após permutação: {a}")
print(f"Valor de b após permutação: {b}")
```
